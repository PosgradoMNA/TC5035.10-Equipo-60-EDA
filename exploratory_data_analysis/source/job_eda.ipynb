```python
def extract_string_range_data(df : pd.DataFrame, pattern : str, column : str, min_column : str, max_column : str):
    """
    """
    df[[min_column, max_column]] = (
    df[column]
    .str.extract(pattern)
    .astype(int)
)
```

```python
job_data[["min_experience_years", "max_experience_years"]] = (
    job_data["Experience"]
    .str.extract(r"(\d+)\s*to\s*(\d+)")
    .astype(int)
)
```

```python
job_data[["min_salary", "max_salary"]] = (
    job_data["Salary Range"]
    .str.extract(r"\$(\d+)K-\$(\d+)K")
    .astype(int)
)

job_data["min_salary"] = job_data["min_salary"] * 1000
job_data["max_salary"] = job_data["max_salary"] * 1000
```

# Extraccion de datos de pdf
---

```python
import pymupdf4llm
import os
from pypdf import PdfReader
from unstructured.partition.auto import partition
```

```python
FILE_PATH = "../files/input_cv"
FILE_NAME = "cv_test.pdf"
```

## pymupdf4llm — The Markdown Generator


```python
markdown = pymupdf4llm.to_markdown(os.path.join(FILE_PATH, FILE_NAME))
print(markdown)
```

What I got: Clean markdown output in 0.14 seconds with proper headings and table formatting. Surprisingly good results. Makes errors that can be cleaned due to the accents. The results are created in markdown.


Good for: Content systems, documentation processing, when you need structured text that preserves hierarchy.


Consider if: You’re dealing with complex multi-column layouts that might get scrambled.

## pypdf — The Reliable Default


```python
reader = PdfReader(os.path.join(FILE_PATH, FILE_NAME))
text = "\n".join(p.extract_text() for p in reader.pages)
print(text)
```

What I got: Solid text extraction with occasional spacing quirks. Works everywhere, no C dependencies. Makes mistakes with accents, clear text and good structure


Good for: Lambda functions, containerised apps, environments where you can’t compile extensions.


Consider if: Text fidelity is critical for your downstream processing

## unstructured — The Semantic Chunker


```python
blocks = partition(filename = os.path.join(FILE_PATH, FILE_NAME))
for block in blocks:
    print(f"{block.category}: {block.text}")
```

What I got: Semantically labelled chunks (Title, NarrativeText, etc.) in 1.11 seconds. Perfect for downstream processing. Separates everything with Title, Headers, NarrativeText and Footers with the ability to specify the language.


Good for: RAG systems, document analysis, when you need meaningful content boundaries for embedding
